```python
# csr = compressed sparse row
# csc = compresed sparse column
from scipy.sparse import csr_matrix 
arr = np.array([0, 0, 0, 0, 0, 1, 1, 0, 2])
print(csr_matrix(arr))
```

```python
arr = np.array([[0, 0, 0], [0, 0, 1], [1, 0, 2]])
print(csr_matrix(arr))
print(csr_matrix(arr).data) # .data used to view the non-zero values
```

```python
new_arr = csr_matrix(arr).tocsc()
print(new_arr)
```

```python
# question 1
obj = plt.figure()
ax = obj.add_axes([0,0,1,1])
x = np.arange(0,100)
y = x*2
ax.plot(x,y)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('title')
```

```python
# question 3
company_df = pd.read_csv('company_sales_data.csv')
print(company_df)
```
